```python
import twinkle
import numpy as np
```

```python
Nsrcs = 1024
device_num = 0
N_stream = 1
RelTol = 1e-4

LittleStar = twinkle.Twinkle(Nsrcs,device_num,N_stream, RelTol)
```

```python
Nsrcs = 1024
ss = 0.5
qq = 1e-6
rho = 1e-4
RelTol = 1e-4
xs = np.empty(Nsrcs)
ys = np.empty(Nsrcs)
for i in range(Nsrcs):
    xs[i] = (i/Nsrcs) * 1e-4 - 1.5001 + 0.11*1e-3
    ys[i] = -0.0035
Mag = np.empty(Nsrcs)

```

```python
LittleStar.set_params(ss,qq,rho,xs,ys)
LittleStar.run()
LittleStar.return_mag_to(Mag)
```

```python
LD_a = 1.       # linear limb darkening effect

RelTol = 1e-6
MagLD = np.empty(Nsrcs)
LittleStar.set_params(ss,qq,rho,xs,ys)
LittleStar.runLD( LD_a )
LittleStar.return_mag_to(MagLD)
```

```python
import matplotlib.pyplot as plt
plt.plot(Mag)
plt.plot(MagLD)
```
