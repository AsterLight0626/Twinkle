```python
import twinkle
import numpy as np
```

```python
Nsrcs = 1024
device_num = 0
N_stream = 1

LittleStar = twinkle.Twinkle(Nsrcs,device_num,N_stream)
```

```python
ss = 0.5
qq = 1e-6
rho = 1e-4
RelTol = 1e-4
xs = np.empty(Nsrcs)
ys = np.empty(Nsrcs)
for i in range(Nsrcs):
    xs[i] = (i/Nsrcs) * 1e-4 - 1.5001 + 0.11*1e-3
    ys[i] = -0.0035
Mag = np.empty(Nsrcs)

```

```python
LittleStar.set_params(ss,qq,rho,RelTol,xs,ys)
LittleStar.run()
LittleStar.return_mag_to(Mag)
```
